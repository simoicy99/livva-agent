```python
import os
import pandas as pd

# Define the data directory relative to this notebook
data_dir = os.path.join(os.pardir, 'data')

# List all CSV files in the data directory
csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

# Load each CSV into a DataFrame and store in a dictionary
dataframes = {}
for csv_filename in csv_files:
    path = os.path.join(data_dir, csv_filename)
    df_name = os.path.splitext(csv_filename)[0]
    dataframes[df_name] = pd.read_csv(path)

# Display the names and shapes of the loaded DataFrames
for name, df in dataframes.items():
    print(f"{name}: {df.shape}")

```
